```python
import pandas as pd
import numpy as np
import os
import math
from itertools import product
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, matthews_corrcoef, roc_curve
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from joblib import Parallel, delayed

# Function to calculate perplexity for a sequence based on n-grams
def calculate_perplexity(seq, n):
    if len(seq) < n:
        return np.nan
    ngrams = [seq[i:i+n] for i in range(len(seq) - n + 1)]
    counts = {ngram: ngrams.count(ngram) for ngram in set(ngrams)}
    total = sum(counts.values())
    
    # Calculate probabilities
    probabilities = {ngram: count / total for ngram, count in counts.items() if count > 0}
    
    # Calculate entropy
    entropy = -sum(p * math.log2(p) for p in probabilities.values())
    
    # Calculate perplexity
    perplexity = 2 ** entropy
    return perplexity

# Function to transform a sequence using perplexity calculation over different n-grams and window sizes
def transform_sequence_with_perplexity(sequence, window_size=10, n=1, overlapping=False):
    step_size = 1 if overlapping else window_size
    perplexity_values = [calculate_perplexity(sequence[i:i+window_size], n) for i in range(0, len(sequence) - window_size + 1, step_size)]
    return np.array(perplexity_values)

# Load and transform data from positive and negative file content
def load_and_transform_data(file_content, label, window_size=10, n=1, overlapping=False):
    sequences = file_content.strip().split('\n')[:300]  # Limit to 300 sequences
    transformed_data = [transform_sequence_with_perplexity(seq, window_size, n, overlapping) for seq in sequences if len(seq) >= window_size]
    df = pd.DataFrame(transformed_data)
    
    # Filter out rows with NaN values
    df = df.dropna()
    
    # Add the label column
    df['label'] = label
    
    return df

# Extract sequences from given positions
def extract_sequences(filename, start_pos, end_pos):
    with open(filename, 'r') as file:
        return "\n".join(line[start_pos-1:end_pos] for line in file.read().split('\n')[:300] if len(line) >= end_pos)

# Perform model training and evaluation
def train_and_evaluate_model(model, param_grid, X_train, y_train, X_test, y_test, model_name, file_name, window_size, n, overlapping):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    auc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1]) if hasattr(best_model, 'predict_proba') else None
    mcc = matthews_corrcoef(y_test, y_pred)

    return {
        'File': file_name,
        'Window Size': window_size,
        'N-gram': n,
        'Overlapping': overlapping,
        'Classifier': model_name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1 Score': f1_score(y_test, y_pred, zero_division=0),
        'AUC': auc,
        'MCC': mcc
    }

# Main processing function
def process_files(files, output_csv='classification_results_perplexity_ngram.csv'):
    results = []
    
    window_sizes = [5, 10, 15]  # Different window sizes
    ngram_lengths = [1, 2, 3]  # Mononucleotide, Dinucleotide, Trinucleotide
    overlap_options = [True, False]  # Overlapping and non-overlapping

    models = {
        'Random Forest': (RandomForestClassifier(random_state=101), {'n_estimators': [100, 200], 'max_depth': [5, 10]}),
        'SVM': (SVC(probability=True, random_state=101), {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}),
        'Logistic Regression': (LogisticRegression(random_state=101), {'C': [0.1, 1, 10]}),
        'Naive Bayes': (GaussianNB(), {}),
        'K-NN': (KNeighborsClassifier(), {'n_neighbors': [3, 5, 7]}),
        'Gradient Boosting': (GradientBoostingClassifier(random_state=101), {'n_estimators': [100, 200], 'learning_rate': [0.05, 0.1, 0.2]}),
        'AdaBoost': (AdaBoostClassifier(random_state=101), {'n_estimators': [50, 100]}),
        'Decision Tree': (DecisionTreeClassifier(random_state=101), {'max_depth': [3, 5, 7]}),
        'XGBoost': (XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=101), {'n_estimators': [100, 200], 'max_depth': [3, 5, 7], 'learning_rate': [0.05, 0.1, 0.2]}),
        'LightGBM': (LGBMClassifier(random_state=101), {'n_estimators': [100, 200], 'num_leaves': [31, 62]}),
        'CatBoost': (CatBoostClassifier(verbose=0, random_state=101), {'iterations': [100, 200], 'depth': [4, 6]})
    }

    for file_name in files:
        print(f"Processing {file_name} with various window sizes, n-grams, and overlapping options")
        neg_content = extract_sequences(file_name, 1, 300)
        pos_content = extract_sequences(file_name, 5001, 5300)

        for window_size, n, overlapping in product(window_sizes, ngram_lengths, overlap_options):
            pos_data = load_and_transform_data(pos_content, label=1, window_size=window_size, n=n, overlapping=overlapping)
            neg_data = load_and_transform_data(neg_content, label=0, window_size=window_size, n=n, overlapping=overlapping)
            combined_data = pd.concat([pos_data, neg_data], ignore_index=True)

            # Ensure there are no NaNs left
            combined_data = combined_data.dropna()
            if combined_data.empty:
                print(f"Skipping configuration for {file_name}: Window Size {window_size}, N-gram {n}, Overlapping {overlapping} due to no valid data.")
                continue

            X, y = combined_data.drop('label', axis=1), combined_data['label']

            # Double-check and report any NaNs before proceeding
            if X.isnull().any().any():
                print(f"NaN values found in features for {file_name}, Window Size {window_size}, N-gram {n}, Overlapping {overlapping}. Skipping this configuration.")
                continue

            # Scale the data
            scaler = StandardScaler()
            X_scaled = scaler.fit_transform(X)

            X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=101)

            # Parallel model training and evaluation
            model_results = Parallel(n_jobs=-1)(delayed(train_and_evaluate_model)(
                model, param_grid, X_train, y_train, X_test, y_test, model_name, file_name, window_size, n, overlapping
            ) for model_name, (model, param_grid) in models.items())

            results.extend(model_results)
            print(f"Results for {file_name}, Window Size {window_size}, N-gram {n}, Overlapping {overlapping} exported to {output_csv}")

    results_df = pd.DataFrame(results)
    results_df.to_csv(output_csv, index=False)
    print("Results exported to", output_csv)
    print(results_df)

# List of files to process
files_to_process = [
    'Kluy_correct_10001',
    'Pichia_correct_10001',
    'Schizo_correct_10001',
    'Candiada_correct_10001',
    'Arabidopsis_correct_10001',
    'Hg19_2000_10001',
    'Sacharo_correct_10001',
    'MM8_correct_10001',
    'DM3_correct_10001'
]

process_files(files_to_process)
```

```python
import glob
import matplotlib.pyplot as plt
import math

# Function to read sequences from a plain text file
def read_sequences(filename):
    with open(filename, 'r') as file:
        sequences = file.read().splitlines()  # Read all lines as individual sequences
    return sequences

# Function to calculate perplexity for any n-gram
def calculate_perplexity(seq, n):
    if len(seq) < n:
        return float('inf')
    
    ngrams = [seq[i:i+n] for i in range(len(seq) - n + 1)]
    counts = {ngram: ngrams.count(ngram) for ngram in set(ngrams)}
    total = sum(counts.values())
    
    if total == 0:
        return float('inf')
    
    probabilities = {ngram: count / total for ngram, count in counts.items() if count > 0}
    entropy = -sum(p * math.log2(p) for p in probabilities.values())
    perplexity = 2 ** entropy
    return perplexity

# Process multiple sequence files and calculate average perplexity for different n-grams and windows
def process_sequence_files(pattern):
    files = glob.glob(pattern)
    if not files:
        print(f"No files found matching pattern: {pattern}")
        return

    num_files = len(files)
    ngram_lengths = [1, 2, 3]  # Mononucleotide, Dinucleotide, Trinucleotide
    window_sizes = [5, 10, 20]  # Different window sizes
    step_size = 1  # Overlapping by 1 nucleotide

    # Create a grid of subplots
    fig, axs = plt.subplots(num_files, len(ngram_lengths), figsize=(18, 6 * num_files))

    for i, file in enumerate(files):
        sequences = read_sequences(file)

        for j, (n, window_size) in enumerate(zip(ngram_lengths, window_sizes)):
            perplexities = []
            sequence_count = 0

            for sequence in sequences:
                if 'N' in sequence:
                    continue
                
                sequence_len = len(sequence)
                for k in range(0, sequence_len - window_size + 1, step_size):
                    window = sequence[k:k+window_size]
                    
                    # Calculate perplexity for the current n-gram
                    perplexity = calculate_perplexity(window, n)
                    
                    # Store perplexities
                    while len(perplexities) <= k // step_size:
                        perplexities.append([])

                    perplexities[k // step_size].append(perplexity)
                
                sequence_count += 1

            if sequence_count == 0:
                print(f"No valid sequences found in {file}.")
                continue

            # Calculate average perplexities
            average_perplexity = [sum(pos) / len(pos) if pos else 0 for pos in perplexities]

            # Normalize perplexities
            min_perplexity = min(average_perplexity)
            max_perplexity = max(average_perplexity)
            normalized_perplexity = [(p - min_perplexity) / (max_perplexity - min_perplexity) if max_perplexity != min_perplexity else 0 for p in average_perplexity]

            # Plotting
            axs[i, j].plot(range(len(normalized_perplexity)), normalized_perplexity, marker='o', linestyle='-', color='b')
            axs[i, j].set_title(f'{["Mono", "Di", "Tri"][n-1]}nucleotide Perplexity\nWindow {window_size} for {file}')
            axs[i, j].set_xlabel('Position of Nucleotide (Window Index)')
            axs[i, j].set_ylabel('Normalized Perplexity')

    plt.tight_layout()
    plt.show()

# Process all sequence files matching the specified pattern
process_sequence_files("*10001")
```
